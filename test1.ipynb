```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import quandl
```

```python
csv_file1 = 'dataset/Europe_Brent_Spot_Price_FOB.csv'
data = pd.read_csv(csv_file1)
```

```python
data = data.reindex(index=data.index[::-1])
```

```python
plt.ylabel("Crude Oil Prices: Brent - Europe")
data.Value.plot(figsize=(10,5))
```

```python
data['MA3'] = data['Value'].shift(1).rolling(window=3).mean()
data['MA9']= data['Value'].shift(1).rolling(window=9).mean()
```

```python
data = data.dropna()
X = data[['MA3','MA9']]
y = data['Value']
print(len(data))
```

```python
training = 0.7
t = int(training*len(data))

X_train = X[:t]
y_train = y[:t]

X_test = X[t:]
y_test = y[t:]
```

```python
model = LinearRegression().fit(X_train,y_train)
```

```python
predicted_price = model.predict(X_test)
predicted_price = pd.DataFrame(predicted_price,index=y_test.index,columns = ['price'])
predicted_price.plot(figsize=(10,5))
y_test.plot()
plt.legend(['Predicted Price','Actual Price'])
plt.ylabel("Crude Oil Prices: Brent - Europe")
plt.show()
```
